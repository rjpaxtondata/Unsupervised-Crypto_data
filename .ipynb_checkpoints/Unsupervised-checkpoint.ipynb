```python
#import dependencies
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
#read CSV file
d1 = ('crypto_data.csv')
df = pd.read_csv(d1)
df.head()
```

```python
# drop the cryptocurrencies that are not traded
df2 = df[df['IsTrading']!= False]

#     or df.query("IsTrading==True")
#Print the frequency for the IsTraiding variable
df2['IsTrading'].value_counts()
```

```python
#rename first column
df3=df2.rename(columns={'Unnamed: 0': 'Symbol'})
df3.head()
```

```python
# drop missing rows
df3.dropna(inplace=True)
```

```python
#remove coins that have not been mined
df4 = df3[df3['TotalCoinsMined']>0]
df4.head(20)
```

```python
#set index as the symbol
df4.set_index('Symbol', inplace=True)
df4
```

```python
#Drop unused columns
df5 = df4.drop(['CoinName', 'IsTrading'], axis=1)
df5.head()
```

```python
#create dummies for categorical variables
dummies = pd.get_dummies(df5, columns = ['Algorithm', 'ProofType'])
dummies
```

```python
#Standardize the continuous variables

scaled = StandardScaler().fit_transform(dummies)
scaled
```

```python
# Dimension Reduction

#Initialize PCA Model

pca = PCA(n_components=0.90)

#fit the model
crypto_pca = pca.fit_transform(scaled)

#create a pandas dataframe of the principal components
pca_df = pd.DataFrame(crypto_pca)
pca_df
```

```python
# Fetch the explained variance
pca.explained_variance_ratio_
```

```python
#further reduce the dataset dimensions with t-SNE
from sklearn.manifold import TSNE
tsne = TSNE(learning_rate=200)

tsne_features = tsne.fit_transform(pca_df)
```

```python
#Determine the number of TSNE Features
tsne_features.shape
```

```python
#Create a scatter plot by principal component

plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
# Initializing model with K = 1-10
from sklearn.cluster import KMeans
inertia = []
k = list(range(1, 11))

#for loop to identify the best k
for i in k:
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(pca_df)
    inertia.append(km.inertia_)


# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
